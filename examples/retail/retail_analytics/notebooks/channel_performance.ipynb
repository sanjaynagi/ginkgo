```python
channel_metrics = "results/channel_metrics.csv"
report_title = "Retail Channel Performance Notebook"
```

```python
from pathlib import Path

import pandas as pd
```

```python
frame = pd.read_csv(Path(channel_metrics))
summary = frame.groupby("channel", as_index=False).agg(
    revenue=("revenue", "sum"),
    margin=("margin", "sum"),
    orders=("orders", "sum"),
)
summary["avg_revenue_per_order"] = (summary["revenue"] / summary["orders"]).round(2)
print(report_title)
print(summary.to_string(index=False))
```
